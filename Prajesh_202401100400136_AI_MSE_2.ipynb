```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files

# Function to upload dataset using Google Colab file upload
def upload_dataset():
    uploaded = files.upload()
    if uploaded:
        # Assuming the file is a CSV, retrieve the file name
        file_name = next(iter(uploaded))
        return file_name
    else:
        print("No file uploaded.")
        return None

# Load and display the dataset
def load_and_display_data(file_path):
    # Load dataset using pandas
    data = pd.read_csv(file_path)

    # Print the first few rows of the dataset (a preview)
    print("Dataset preview:")
    print(data.head())

    # Show basic information about the dataset (e.g., columns, non-null counts)
    print("\nDataset info:")
    data.info()

    return data

# Visualize dataset (e.g., histograms, correlation heatmap)
def visualize_data(data):
    # Plot histograms for numeric columns
    data.hist(bins=20, figsize=(12, 10))
    plt.tight_layout()
    plt.show()

    # Select only numeric columns for correlation matrix
    numeric_data = data.select_dtypes(include=['number'])

    # Optionally drop 'id' or unnamed columns
    numeric_data = numeric_data.loc[:, ~numeric_data.columns.str.contains('^Unnamed|id', case=False)]

    # Compute correlation matrix
    correlation_matrix = numeric_data.corr()

    # Plot correlation heatmap with better spacing
    plt.figure(figsize=(16, 14))  # Increased size for better layout
    sns.heatmap(
        correlation_matrix,
        annot=True,
        cmap="coolwarm",
        fmt=".2f",
        square=True,
        annot_kws={"size": 8}
    )
    plt.xticks(rotation=45, ha='right', fontsize=10)
    plt.yticks(rotation=0, fontsize=10)
    plt.title("Correlation Heatmap", fontsize=16, pad=20)
    plt.tight_layout()
    plt.show()

def main():
    # Step 1: Upload dataset
    file_path = upload_dataset()
    if not file_path:
        return

    # Step 2: Load and display the dataset
    data = load_and_display_data(file_path)

    # Step 3: Visualize the data
    visualize_data(data)

# Run the main function
main()
```
